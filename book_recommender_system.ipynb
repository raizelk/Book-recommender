# Content Based Recommender system (Author + Genre)

```python
unique_books = unique_books[['title','author_x','description','genres','setting','img']]
```

```python
unique_books.head()
```

```python
unique_books.isnull().sum()
```

```python
unique_books['description'] = unique_books['description'].astype(str)
```

```python
unique_books['description'] = unique_books['description'].apply(lambda x:x.split())
```

```python

unique_books.head()
```

```python
unique_books['author_x'] = unique_books['author_x'].apply(lambda x: [x])
```

```python
unique_books.head()
```

```python
def str_to_list(string):
    cleaned_string = string.strip("[]")
    return [item.strip() for item in cleaned_string.split(',')]
unique_books['genres'] = unique_books['genres'].apply(str_to_list)


```

```python
unique_books['tags'] = unique_books['author_x'] + unique_books['description'] + unique_books['genres'] + unique_books['setting']
```

```python
unique_books.head()
```

```python
unique_books = unique_books[['title','tags','img']]
unique_books.head()
```

```python
unique_books['tags'] = unique_books['tags'].apply(lambda x:" ".join(x))
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
unique_books['tags'] = unique_books['tags'].apply(stem)
```

```python
unique_books['tags'][0]
```

```python
unique_books['tags'] = unique_books['tags'].apply(lambda x:x.lower())
```

```python
unique_books.head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(unique_books['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(book):
    book_index = unique_books[unique_books['title'] == book].index[0]
    distances = similarity[book_index]
    book_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in book_list:
        print(unique_books.iloc[i[0]].title)
```

```python
recommend('Twilight')
```
